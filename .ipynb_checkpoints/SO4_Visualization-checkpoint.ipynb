```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

# Line Plot

```python
plt.plot(years, iran_pop)
plt.show()
```

# Scatter Plot

```python
plt.scatter(years, iran_pop)
plt.show()
```

# Histogram Plot 

```python
city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1547164, 1448075, 1424641, 1249942]
```

```python
plt.hist(city_pop, bins= 'auto',rwidth=0.8)
plt.show()
```

# Pie Chart

```python
plt.figure(figsize= (12,8))
plt.pie(city_pop, labels= city_name, labeldistance=0.5)
plt.show()
```

# Customization Plots 

### Customize Line Plot

```python
plt.figure(figsize=(8,6), dpi= 100)
plt.plot(years, iran_pop)
plt.xlabel('Years', fontsize = 14 )
plt.ylabel('Pop', fontsize = 14)
plt.xticks(ticks= years)
plt.yticks(ticks= iran_pop, 
            labels= ['21M', '28M', '38M', '56M', '66M', '74M', '80M'])
plt.title('Iran\'s \n Population')
plt.show()
```

### Customize Scatter plot

```python
plt.figure(figsize=(8,6), dpi= 100)
Pop_size = np.array(city_pop)/3000
Colors = ['violet', 'green', 'orange', 'tomato', 'blue', 'pink']
plt.scatter(np.arange(6), city_pop, s=Pop_size , c= Colors)
#plt.scatter(np.arange(6), city_pop)
plt.title('Iran\'s most \n Populated city', fontsize = 14, c= 'y')
plt.xlabel('City\'s name', fontsize = 14, c= 'y')
plt.ylabel('Population', fontsize = 14, c = 'y')
plt.xticks(ticks= np.arange(6), labels= city_name, c = 'orange')
plt.yticks(ticks= [1e6, 2e6, 7e6], labels= ['1M', '2M', '7M'], c = 'orange')
plt.text(x=0 , y= 7e6, s= 'iran Capital', fontsize= 14)
plt.margins(0.2)
plt.show()
# plt.savefig("test.pdf", dpi=150)
# plt.savefig("test.png", dpi=150)
```

# Multiple Plot

```python
plt.figure(figsize=(8,6), dpi= 100)
turkey_pop = [27.47, 34.88, 43.98, 53.92, 63.24, 72.33, 79.51]
plt.plot(years, iran_pop, ls = '-', marker = '*', lw = 0.5, mew = 2, mec = 'red')
plt.plot(years, turkey_pop, ls = '--', marker = 'o', lw = 0.8, mew = 2, mec = 'blue')
plt.title('Iran vs Turkey Population', c='y', fontsize= 14)
plt.xlabel('Years', c='y', fontsize= 14)
plt.ylabel('Population', c= 'y', fontsize = 14)
plt.yticks([20, 30, 40, 50, 60, 70, 80], 
           ['20M', '30M', '40M', '50M', '60M', '70M', '80M'], c= 'orange')
plt.xticks(years, c='orange')
plt.legend(['Iran', 'Turkey'], loc= 'best')
plt.annotate('Iran-Iraq War', xy= (1980, 38.67), xytext = (1990, 34.70), fontsize= 14,
             c= 'red', arrowprops = dict(arrowstyle= '-|>', lw = 2 , ls= '--' , color='black'))
plt.show()
```

# Subplot

```python
plt.subplot(1,2,1)
plt.plot(years, iran_pop)
plt.title('Iran\'s Population', c = 'y')

plt.subplot(1,2,2)
plt.plot(years, turkey_pop)
plt.title('Turkey\'s Population', c = 'y')

plt.show()
```

# Seaborn

```python
import pandas as pd 
```

```python
import seaborn as sb 
```

```python
df = pd.read_csv('smartphones.csv')
```

```python
df
```

# Strip Plot 

## stripplot()

```python
sb.stripplot(x = 'OS', y = 'Capacity', data= df, hue= 'OS' ,size= 10, jitter= 1 )

plt.show()
```

## swarmplot()

```python
sb.swarmplot(x = 'OS', y= 'Capacity', hue= 'OS' ,data= df, size = 10)
plt.show()
```

```python
sb.swarmplot(x = 'OS', y= 'Capacity', hue= 'Company' ,data= df, size = 10)
plt.show()
```
