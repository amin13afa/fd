```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import datasets 
```

```python
iris_data = pd.read_csv('iris.data', header= None)
```

```python
names = {0:'s_l',
        1:'s_w',
        2:'p_l',
        3:'p_w',
        4:'class'}
iris_data.rename(columns= names, inplace= True)
```

```python
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
print(iris.DESCR)
```

```python
iris.target_names
```

```python
df = pd.DataFrame(data= iris.data, columns= iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
```

```python
names_2 = {'sepal length (cm)' : 's_l',
          'sepal width (cm)' : 's_w',
          'petal length (cm)' : 'p_l', 
          'petal width (cm)' : 'p_w',
          'target' : 't'}
```

```python
df.rename(columns= names_2, inplace= True)
```

```python

pd.plotting.scatter_matrix(df, c = iris.target , figsize= (12,8))
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 6, metric='minkowski', p = 2)
```

```python
x = iris.data
y = iris.target
```

```python
knn.fit(x,y)
```

```python
x_new = np.array([[5, 3, 1, 0.2]])
#y_new = ??
y_new = knn.predict(x_new)
y_new
```
